```python
import polars as pl
import seaborn as sb
```

```python
df = pl.scan_csv("countries.csv")
```

```python
df.collect()
```

```python
df.select([
    pl.pearson_corr("income", "gdpp").alias("income-gdpp"),
    pl.pearson_corr("income", "total_fer").alias("income-total_fer"),
    pl.pearson_corr("income", "child_mort").alias("income-child_mort"),
    pl.pearson_corr("imports", "exports").alias("imports-exports"),
    pl.pearson_corr("total_fer", "child_mort").alias("total_fer-child_mort"),
    pl.pearson_corr("gdpp", "child_mort").alias("gdpp-child_mort")
]).collect()
```

```python
sb.pairplot(df.select(pl.col(["income", "gdpp", "total_fer", "child_mort", "imports", "exports"])).collect().to_pandas())
```
