```python
df.select([
    pl.spearman_rank_corr("income", "total_fer").alias("income-total_fer"),
    pl.spearman_rank_corr("income", "child_mort").alias("income-child_mort"),
    pl.spearman_rank_corr("gdpp", "child_mort").alias("gdpp-child_mort")
]).collect()
```

What this tells me is that there is absolutely a relationship between the money in the country and fertility of the country. Talking with my fiancée, much of this trend is cause-effect. The antecedent is the ability to preserve a child's life, that is children die less. As these increase the need for multiple children decreases. Then, over the course of time, as societies move towards a western model of civilization, women tend to work more. This is seen with the countries with higher income having $\rho=0.74$ in regards to income vs total fertility rate. As such, the countries with larger child mortality rates are in the most need of medical help.

Another item to look at is the relation between health and GDP per Capita. Health measures how much of the GDP is spent on Healthcare per Capita. Thus the assumed relationship would be linear.

```python
print(df.select([
    pl.pearson_corr("gdpp", "health").alias("gdpp-health-linear"),
    pl.spearman_rank_corr("gdpp", "health").alias("gdpp-health-spearman")
]).collect())
sb.scatterplot(x="gdpp", y="health", data=df.collect()).set(title="Health vs GDPP")
```

However, as we can see there is no relation to it. In conjunction with this would be the graph for health vs Child Mortality vs Total Fertility Rate.

```python
plt = sb.PairGrid(df.collect().to_pandas(), x_vars=["child_mort", "total_fer"], y_vars=["health"])
plt.fig.suptitle("Health vs Child Mortality vs Total Fertility Rate")
plt.fig.subplots_adjust(top=0.9)
plt.map_diag(sb.scatterplot)
plt.map_offdiag(sb.scatterplot)
print(plt)
```

And there is absolutely no real correlation between these. So giving money outright is not a solution to healthcare, rather it is providing worthwhile healthcare.
